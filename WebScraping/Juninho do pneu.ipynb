```python
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
url = "https://www.camara.leg.br/deputados/204457"
```

```python
option = Options()
option.headless = True
driver = webdriver.Chrome("C:\WebDriver/chromedriver.exe")
driver.get(url)
driver.find_element_by_xpath("//div[@class='mm-page mm-slideout']//div[@class='js-mmenu-container']//div[@class='container-blocos-deputado']")
```

```python
element = driver.find_element_by_xpath(
    "//div[@class='mm-page mm-slideout']//div[@class='js-mmenu-container']//div[@class='container-blocos-deputado']")
```

```python
html_content = element.get_attribute("outerHTML")
```

```python
print(html_content)
```

```python
soup = BeautifulSoup(html_content,'html.parser')
```

```python
print(soup)
```

```python
type(soup)
```

```python
mylist = soup.find_all('h2', class_='nome-deputado')
```

```python
type(mylist)
```

```python
nome=[]
for x in mylist:
    nome.append(str(x))
```
