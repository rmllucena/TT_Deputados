```python
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from pandas import DataFrame
```

```python
linktransparencia = [["Glauber Braga","152605","2001"], ["Tabata Amaral","204534","3208"], ["Marcelo Freixo","76874","3222"]]
print(len(linktransparencia))
```

```python
tabelafinal = []
mesatual = 12

for link in range(len(linktransparencia)):
    linktransp = "https://www.camara.leg.br/deputados/{}".format(linktransparencia[link][1])


    #PEGAR O NOME


    option = webdriver.ChromeOptions()
    option.add_argument('headless') 
    option.add_argument("enable-automation");
    option.add_argument("--window-size=1920,1080");
    option.add_argument("--no-sandbox");
    option.add_argument("--disable-extensions");
    option.add_argument("--dns-prefetch-disable");
    option.add_argument("--disable-gpu");
    driver = webdriver.Chrome(r"C:\Users\Romulo\Programação\Python\Scripts\Twitter\TT_Deputados\WebScraping\Driver\chromedriver.exe",options=option)
    driver.get(linktransp)
    driver.find_element_by_xpath(
                "//div[@class='mm-page mm-slideout']//div[@class='js-mmenu-container']//div[@class='container-blocos-deputado']")
    element = driver.find_element_by_xpath(
                "//div[@class='mm-page mm-slideout']//div[@class='js-mmenu-container']//div[@class='container-blocos-deputado']")

    html_content = element.get_attribute("outerHTML")

    soup = BeautifulSoup(html_content,'html.parser')

    mylist = soup.find_all('h2', class_='nome-deputado')

    nome=[]
    for x in mylist:
        nome.append(str(x))
    dados = []
    mylist = soup.find_all('li')
    for item in mylist:
        dados.append(item.text)





    dadosjuntos = [[dados[0].split(":",1)[1],
                            dados[1].split(":",1)[1],
                            dados[2].split(":",1)[1],
                            " ".join(dados[3].split(":",1)[1].replace('\n', '').split()),
                            " ".join(dados[4].split(":",1)[1].replace('\n', '').split()),
                            " ".join(dados[5].split(":",1)[1].replace('\n', '').split())]]

    #print("printar dadosjuntos")
    print("Link {}".format(link))
    driver.close()

    
    
    
    
    
    #PEGAR ORÇAMENTO
    
    ano = 21 
    for depcode in range(len(linktransparencia)):
        mesespecifico = 1

        while mesespecifico < mesatual:
            linkOrçamento = 'https://www.camara.leg.br/cota-parlamentar/sumarizado?nuDeputadoId={}&dataInicio={}/20{}&dataFim={}/20{}&despesa=&nomeHospede=&nomePassageiro=&nomeFornecedor=&cnpjFornecedor=&numDocumento=&sguf=&filtroNivel1=1&filtroNivel2=2&filtroNivel3=3'.format(linktransparencia[link][2],mesespecifico, ano, mesespecifico, ano)
            
            
            option = webdriver.ChromeOptions()
            option.add_argument('headless') 
            option.add_argument("enable-automation");
            option.add_argument("--window-size=1920,1080");
            option.add_argument("--no-sandbox");
            option.add_argument("--disable-extensions");
            option.add_argument("--dns-prefetch-disable");
            option.add_argument("--disable-gpu");
            driver = webdriver.Chrome(r"C:\Users\Romulo\Programação\Python\Scripts\Twitter\TT_Deputados\WebScraping\Driver\chromedriver.exe",options=option)
            driver.get(linkOrçamento)
            driver.find_element_by_xpath(
                        "//div[@class='coluna-centro col-md-9']")
            elementOrçamento = driver.find_element_by_xpath(
                        "//div[@class='coluna-centro col-md-9']")

            html_contentOrçamento = elementOrçamento.get_attribute("outerHTML")

            soupOrçamento = BeautifulSoup(html_contentOrçamento,'html.parser')

            Contagem = 0
            tabeladados = []

            while True:
                try:
                    TipoDespesa = soupOrçamento.find(id="linkSumarizado{}".format(Contagem))
                    quotes = [quote for quote in TipoDespesa]
                    Despesa = soupOrçamento.find(id="sumarizado{}".format(Contagem))
                    quotes2 = [quote for quote in Despesa]
                    tabeladados.append([quotes[0].replace('\n', '').replace('\t', ''),quotes2[0].replace('\n', '').replace('\t', '')])

                    Contagem += 1

                except:
                    break

            mes = soupOrçamento.find_all("h3",class_= "header")

            dadomes =[]
            listmes = [quote for quote in mes]

            for x in listmes:
                 dadomes.append(str(listmes[1]))

            #print(dadomes[0][28:-5])
            #print(tabeladados)
            driver.close()
            mesespecifico += 1




            for x in range(len(tabeladados)):
                tabelafinal.append([linktransparencia[link][0],dadosjuntos[0][0],dadosjuntos[0][1],dadosjuntos[0][2],dadosjuntos[0][3],
                                                    dadosjuntos[0][4],dadosjuntos[0][5],dadomes[0][28:-5],
                                    tabeladados[x][0]," ".join(tabeladados[x][1].replace('R$', '').replace('.', '').split())])

            #print(tabelafinal)
            print([linktransparencia[link][0],dadosjuntos[0][0],dadomes[0][28:-5]])
        else:
            break


```
